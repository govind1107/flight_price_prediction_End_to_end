# Duration

```python
def get_time(x):
    x = x.split(' ')
    hours = 0
    mins = 0
    if len(x)==1:
        x = x[0]
        if x[-1]=='h':
            hours = x[:-1]
        else:
            mins = x[:-1]
    else:
        hours = x[0][:-1]
        mins = x[1][:-1]
    return hours,mins
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```
